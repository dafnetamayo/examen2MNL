```python
def normalize_minmax(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
def plot_decomposition(df_series, series_id, period=24):
    """
    Realiza la descomposición de una serie temporal en tendencia, estacionalidad y residuales.

    Parámetros:
    - df_series: DataFrame que contiene todas las series transpuestas (series en columnas, fechas en filas)
    - series_id: ID de la serie que se desea analizar (ej. 'H110')
    - period: Frecuencia estacional, por defecto 24 (horas para series horarias diarias)

    Retorna:
    - Gráfica con la descomposición
    """

    start_date = pd.to_datetime("01-07-15 12:00", format="%d-%m-%y %H:%M")
    serie = df_series[series_id].copy()

    # Generar el índice de tiempo
    index = pd.date_range(start=start_date, periods=len(serie), freq="H")
    serie.index = index

    # Crear DataFrame
    df_serie = pd.DataFrame({
        "fecha": index,
        "valor": serie.values
    })

    # Eliminar valores faltantes
    df_serie = df_serie.dropna(subset=['valor'])

    # Normalizar la serie
    df_serie["valor"] = normalize_minmax(df_serie["valor"])

    # Descomposición
    decomposition = seasonal_decompose(df_serie["valor"], model='additive', period=period)

    # Gráficas
    plt.figure(figsize=(12, 8))

    plt.subplot(411)
    plt.plot(df_serie["fecha"], decomposition.observed, label='Original (Normalizada)')
    plt.legend(loc='upper left')

    plt.subplot(412)
    plt.plot(df_serie["fecha"], decomposition.trend, label='Tendencia', color='orange')
    plt.legend(loc='upper left')

    plt.subplot(413)
    plt.plot(df_serie["fecha"], decomposition.seasonal, label='Estacionalidad', color='green')
    plt.legend(loc='upper left')

    plt.subplot(414)
    plt.plot(df_serie["fecha"], decomposition.resid, label='Residuo', color='red')
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.suptitle(f"Descomposición de la serie {series_id} (Normalizada)", y=1.02)
    plt.show()
```

```python
def split_and_plot_series(df_series, df_meta, serie_id):
    meta_serie = df_meta[df_meta["M4id"] == serie_id]
    start_date = pd.to_datetime(meta_serie["StartingDate"].values[0], format="%d-%m-%y %H:%M")
    horizon = int(meta_serie["Horizon"].values[0])

    # Extraer la serie
    serie = df_series[serie_id].dropna().copy()
    index = pd.date_range(start=start_date, periods=len(serie), freq="H")
    serie.index = index

    # Separar el horizonte final
    serie_main = serie.iloc[:-horizon]
    serie_horizon = serie.iloc[-horizon:]

    # División 70/15/15
    n = len(serie_main)
    train_size = int(n * 0.7)
    val_size = int(n * 0.15)
    test_size = n - train_size - val_size  # evita redondeos incorrectos

    serie_train = serie_main.iloc[:train_size]
    serie_val = serie_main.iloc[train_size:train_size + val_size]
    serie_test = serie_main.iloc[train_size + val_size:]

    # Graficar
    plt.figure(figsize=(14, 5))
    plt.plot(serie_train.index, serie_train, label='Entrenamiento (70%)')
    plt.plot(serie_val.index, serie_val, label='Validación (15%)')
    plt.plot(serie_test.index, serie_test, label='Prueba (15%)')
    plt.plot(serie_horizon.index, serie_horizon, label=f'Horizon Final ({horizon})', linestyle='--')

    plt.title(f"División de la Serie {serie_id}")
    plt.xlabel("Fecha")
    plt.ylabel("Valor")
    plt.legend()
    plt.tight_layout()
    plt.grid(True)
    plt.show()

    return serie_train, serie_val, serie_test, serie_horizon
```

```python
def normalize_train_val_test(train, val, test):
    min_val = train.min()
    max_val = train.max()
    train_norm = (train - min_val) / (max_val - min_val)
    val_norm = (val - min_val) / (max_val - min_val)
    test_norm = (test - min_val) / (max_val - min_val)
    return train_norm, val_norm, test_norm, min_val, max_val
```

```python
def split_univariate_sequence(sequence, n_steps):
    X, y = [], []
    for i in range(len(sequence) - n_steps):
        seq_x = sequence[i:i + n_steps]
        seq_y = sequence[i + n_steps]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

## MLP 

```python
def build_mlp_1(n_steps):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps,)),
        keras.layers.Dense(100, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model
```

```python
def build_mlp_2(n_steps):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps,)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model
```

```python
def build_mlp_3(n_steps):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps,)),
        keras.layers.Dense(256, activation='tanh'),
        keras.layers.Dense(128, activation='tanh'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='rmsprop',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model
```

```python
def train_and_evaluate_models(X_train, y_train, X_val, y_val, X_test, y_test, n_steps, min_val, max_val):
    models = {
        "MLP_1_Baseline": build_mlp_1(n_steps),
        "MLP_2_Regularized": build_mlp_2(n_steps),
        "MLP_3_Tanh_RMSProp": build_mlp_3(n_steps)
    }
    
    results = {}

    for name, model in models.items():
        print(f"\nEntrenando {name}...")
        history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                            epochs=50, batch_size=32, verbose=0)

        loss, mae, mape, mse = model.evaluate(X_test, y_test, verbose=0)

        results[name] = {
            "model": model,
            "history": history,
            "MAE": mae,
            "MAPE": mape,
            "MSE": mse
        }

        print(f"{name} - MAE: {mae:.4f}, MAPE: {mape:.2f}%, MSE: {mse:.4f}")
    
    return results
```

```python
def plot_results(results, X_test, y_test, min_val, max_val):
    plt.figure(figsize=(12, 5))
    for name, data in results.items():
        preds = data["model"].predict(X_test).flatten()
        preds_denorm = preds * (max_val - min_val) + min_val
        y_test_denorm = y_test * (max_val - min_val) + min_val
        plt.plot(y_test_denorm[:100], label='Real' if name == "MLP_1_Baseline" else "", color='black')
        plt.plot(preds_denorm[:100], label=name, alpha=0.7)
    plt.title("Comparación de Predicciones")
    plt.legend()
    plt.tight_layout()
    plt.grid(True)
    plt.show()
```

```python
results = train_and_evaluate_models(X_train, y_train, X_val, y_val, X_test, y_test, n_steps, min_val, max_val)
plot_results(results, X_test, y_test, min_val, max_val)
```

```python
def plot_training_loss(results):
    plt.figure(figsize=(14, 6))
    for name, data in results.items():
        history = data["history"]
        plt.plot(history.history["loss"], label=f'{name} - Train')
        plt.plot(history.history["val_loss"], linestyle='--', label=f'{name} - Val')
    plt.title("Función de pérdida (MAE) durante el entrenamiento")
    plt.xlabel("Epoch")
    plt.ylabel("MAE Loss")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_training_loss(results)

```

```python
def predict_future_steps(test_norm, n_steps, results, min_val, max_val,
                         serie_train, serie_val, serie_test, future_horizon=48):
    """
    Realiza predicción a futuro paso a paso (auto-regresiva) con el mejor modelo (según MAE).
    
    Parámetros:
    - test_norm: Serie de prueba normalizada.
    - n_steps: Tamaño de la ventana de entrada.
    - results: Diccionario con los modelos y métricas, como el devuelto por train_and_evaluate_models().
    - min_val, max_val: Valores usados para desnormalizar.
    - serie_train, serie_val, serie_test: Series originales desnormalizadas.
    - future_horizon: Número de pasos a predecir (por defecto 48).
    
    Retorna:
    - predictions_denorm: Predicciones desnormalizadas.
    """

    # Tomar la última ventana conocida
    last_window = test_norm[-n_steps:]
    input_future = np.array(last_window).reshape(1, -1)

    # Escoger el mejor modelo según MAE
    best_model_key = min(results, key=lambda x: results[x]['MAE'])
    best_model = results[best_model_key]['model']

    # Predicciones auto-regresivas
    predictions = []
    current_input = input_future.copy()

    for _ in range(future_horizon):
        next_pred = best_model.predict(current_input, verbose=0)[0][0]
        predictions.append(next_pred)
        current_input = np.append(current_input[:, 1:], [[next_pred]], axis=1)

    # Desnormalizar
    predictions_denorm = np.array(predictions) * (max_val - min_val) + min_val

    # Fechas futuras
    time_steps = pd.date_range(start=serie_test.index[-1] + pd.Timedelta(hours=1),
                               periods=future_horizon, freq='H')

    # Graficar
    plt.figure(figsize=(12, 5))
    plt.plot(serie_train.index, serie_train, label='Entrenamiento')
    plt.plot(serie_val.index, serie_val, label='Validación')
    plt.plot(serie_test.index, serie_test, label='Prueba')
    plt.plot(time_steps, predictions_denorm, label='Predicción 48 horas', color='red', linestyle='--')
    plt.xlabel("Fecha")
    plt.ylabel("Valor")
    plt.title(f"Predicción de las próximas {future_horizon} horas - Mejor Modelo: {best_model_key}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return predictions_denorm
```

Por otro lado, los modelos MLP_2_Regularized y MLP_3_Tanh_RMSProp obtuvieron errores notablemente más altos. En particular, el uso de Dropout y BatchNormalization en MLP_2, así como la activación tanh y el optimizador RMSProp en MLP_3, pudieron haber dificultado el aprendizaje efectivo, sobre todo en una serie relativamente estable. Esto evidencia que, en ciertos casos, una arquitectura simple pero bien ajustada puede superar modelos más sofisticados, siempre que esté bien alineada con la naturaleza de los datos.

```python

n_features = 1  

# Asegúrate de que los datos estén en formato float32
X_train = X_train.astype(np.float32)
X_val   = X_val.astype(np.float32)
X_test  = X_test.astype(np.float32)

# Convertir a formato 3D requerido por CNN: (samples, steps, features)
X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
X_val_cnn   = X_val.reshape((X_val.shape[0], X_val.shape[1], n_features))
X_test_cnn  = X_test.reshape((X_test.shape[0], X_test.shape[1], n_features))


```

```python
def build_cnn_1(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps, n_features)),  # 👈 Esto fuerza la forma
        keras.layers.Conv1D(64, kernel_size=2, activation='relu'),
        keras.layers.MaxPooling1D(),
        keras.layers.Flatten(),
        keras.layers.Dense(50, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model

```

```python
def build_cnn_2(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps, n_features)),
        keras.layers.Conv1D(128, kernel_size=3, activation='relu'),
        keras.layers.Conv1D(64, kernel_size=2, activation='relu'),
        keras.layers.MaxPooling1D(),
        keras.layers.Dropout(0.3),
        keras.layers.Flatten(),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model

```

```python
def build_cnn_3(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps, n_features)),
        keras.layers.Conv1D(64, kernel_size=2, activation='tanh'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling1D(),
        keras.layers.Flatten(),
        keras.layers.Dense(128, activation='tanh'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='rmsprop',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model
```

```python
def train_and_evaluate_cnn_models(X_train, y_train, X_val, y_val, X_test, y_test, n_steps, n_features, min_val, max_val):
    models = {
        "CNN_1_Baseline": build_cnn_1(n_steps, n_features),
        "CNN_2_Deep_Dropout": build_cnn_2(n_steps, n_features),
        "CNN_3_BatchNorm_Tanh": build_cnn_3(n_steps, n_features)
    }

    results = {}

    for name, model in models.items():
        print(f"\nEntrenando {name}...")
        history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                            epochs=50, batch_size=32, verbose=0)

        loss, mae, mape, mse = model.evaluate(X_test, y_test, verbose=0)

        results[name] = {
            "model": model,
            "history": history,
            "MAE": mae,
            "MAPE": mape,
            "MSE": mse
        }

        print(f"{name} - MAE: {mae:.4f}, MAPE: {mape:.2f}%, MSE: {mse:.4f}")
    
    return results
```

```python
results_cnn = train_and_evaluate_cnn_models(
    X_train_cnn, y_train,
    X_val_cnn, y_val,
    X_test_cnn, y_test,
    n_steps, n_features,
    min_val, max_val
)



```

```python
def predict_future_steps_cnn(test_norm, n_steps, results, min_val, max_val,
                              serie_train, serie_val, serie_test, future_horizon=48):
    """
    Predicción paso a paso (auto-regresiva) para CNN univariada.
    
    Parámetros:
    - test_norm: Array 1D normalizado (últimos datos de test).
    - n_steps: Tamaño de ventana.
    - results: Diccionario de modelos con sus métricas (como results_cnn).
    - min_val, max_val: Valores para desnormalizar.
    - serie_train, serie_val, serie_test: Series originales (pandas Series).
    - future_horizon: Número de pasos a predecir (por defecto 48).
    
    Retorna:
    - predictions_denorm: Predicciones desnormalizadas.
    """

    # Selección del mejor modelo
    best_model_key = min(results, key=lambda x: results[x]['MAE'])
    best_model = results[best_model_key]['model']
    print(f"Usando el mejor modelo: {best_model_key}")

    # Preparar la última ventana conocida
    last_window = test_norm[-n_steps:]
    current_input = np.array(last_window, dtype=np.float32).reshape((1, n_steps, 1))

    predictions = []

    for step in range(future_horizon):
        try:
            # Predecir siguiente valor
            next_pred = best_model.predict(current_input, verbose=0)
            next_value = float(next_pred[0][0])
            predictions.append(next_value)

            # Actualizar ventana para la siguiente iteración
            current_input = np.append(current_input[:, 1:, :], [[[next_value]]], axis=1)

        except Exception as e:
            print(f"❌ Error al predecir en paso {step}: {e}")
            break

    if len(predictions) == 0:
        print("❌ No se generaron predicciones.")
        return []

    # Desnormalizar predicciones
    predictions_denorm = np.array(predictions) * (max_val - min_val) + min_val

    # Crear fechas futuras
    time_steps = pd.date_range(start=serie_test.index[-1] + pd.Timedelta(hours=1),
                               periods=future_horizon, freq='h')  # 'h' en minúscula

    # Graficar resultado
    plt.figure(figsize=(12, 5))
    plt.plot(serie_train.index, serie_train, label='Entrenamiento')
    plt.plot(serie_val.index, serie_val, label='Validación')
    plt.plot(serie_test.index, serie_test, label='Prueba')
    plt.plot(time_steps, predictions_denorm, label='Predicción 48h (CNN)', color='red', linestyle='--')
    plt.xlabel("Fecha")
    plt.ylabel("Valor")
    plt.title(f"Predicción futura de {future_horizon} horas - Mejor modelo: {best_model_key}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return predictions_denorm
```
