```python
def normalize_minmax(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
def plot_decomposition(df_series, series_id, period=24):
    start_date = pd.to_datetime("01-07-15 12:00", format="%d-%m-%y %H:%M")
    serie = df_series[series_id].copy()

    # Generar el índice de tiempo
    index = pd.date_range(start=start_date, periods=len(serie), freq="H")
    serie.index = index

    # Crear DataFrame
    df_serie = pd.DataFrame({
        "fecha": index,
        "valor": serie.values
    })

    # Eliminar valores faltantes
    df_serie = df_serie.dropna(subset=['valor'])

    # Normalizar la serie
    df_serie["valor"] = normalize_minmax(df_serie["valor"])

    # Descomposición
    decomposition = seasonal_decompose(df_serie["valor"], model='additive', period=period)

    # Gráficas
    plt.figure(figsize=(12, 8))

    plt.subplot(411)
    plt.plot(df_serie["fecha"], decomposition.observed, label='Original (Normalizada)')
    plt.legend(loc='upper left')

    plt.subplot(412)
    plt.plot(df_serie["fecha"], decomposition.trend, label='Tendencia', color='orange')
    plt.legend(loc='upper left')

    plt.subplot(413)
    plt.plot(df_serie["fecha"], decomposition.seasonal, label='Estacionalidad', color='green')
    plt.legend(loc='upper left')

    plt.subplot(414)
    plt.plot(df_serie["fecha"], decomposition.resid, label='Residuo', color='red')
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.suptitle(f"Descomposición de la serie {series_id} (Normalizada)", y=1.02)
    plt.show()
```

```python
def split_and_plot_series(df_series, df_meta, serie_id):
    meta_serie = df_meta[df_meta["M4id"] == serie_id]
    start_date = pd.to_datetime(meta_serie["StartingDate"].values[0], format="%d-%m-%y %H:%M")
    horizon = int(meta_serie["Horizon"].values[0])

    # Extraer la serie
    serie = df_series[serie_id].dropna().copy()
    index = pd.date_range(start=start_date, periods=len(serie), freq="H")
    serie.index = index

    # Separar el horizonte final
    serie_main = serie.iloc[:-horizon]
    serie_horizon = serie.iloc[-horizon:]

    # División 70/15/15
    n = len(serie_main)
    train_size = int(n * 0.7)
    val_size = int(n * 0.15)
    # test_size se calcula para evitar redondeos incorrectos
    serie_train = serie_main.iloc[:train_size]
    serie_val = serie_main.iloc[train_size:train_size + val_size]
    serie_test = serie_main.iloc[train_size + val_size:]

    # Graficar la división de la serie
    plt.figure(figsize=(14, 5))
    plt.plot(serie_train.index, serie_train, label='Entrenamiento (70%)')
    plt.plot(serie_val.index, serie_val, label='Validación (15%)')
    plt.plot(serie_test.index, serie_test, label='Prueba (15%)')
    plt.plot(serie_horizon.index, serie_horizon, label=f'Horizon Final ({horizon})', linestyle='--')
    plt.title(f"División de la Serie {serie_id}")
    plt.xlabel("Fecha")
    plt.ylabel("Valor")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return serie_train, serie_val, serie_test, serie_horizon
```

```python
def normalize_train_val_test(train, val, test):
    min_val = train.min()
    max_val = train.max()
    train_norm = (train - min_val) / (max_val - min_val)
    val_norm = (val - min_val) / (max_val - min_val)
    test_norm = (test - min_val) / (max_val - min_val)
    return train_norm, val_norm, test_norm, min_val, max_val
```

```python
def split_univariate_sequence(sequence, n_steps):
    X, y = [], []
    for i in range(len(sequence) - n_steps):
        seq_x = sequence[i:i + n_steps]
        seq_y = sequence[i + n_steps]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
serie_id = "H110"
serie_train, serie_val, serie_test, serie_horizon = split_and_plot_series(df_t, df_meta, serie_id)
train_norm, val_norm, test_norm, min_val, max_val = normalize_train_val_test(
    serie_train.values, serie_val.values, serie_test.values
)
```

## MLP 

```python
n_steps = 24
X_train, y_train = split_univariate_sequence(train_norm, n_steps)
X_val, y_val = split_univariate_sequence(val_norm, n_steps)
X_test, y_test = split_univariate_sequence(test_norm, n_steps)
```

```python
def build_mlp_1(n_steps):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps,)),
        keras.layers.Dense(100, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model

def build_mlp_2(n_steps):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps,)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),  
        keras.layers.Dense(64, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model




def build_mlp_3(n_steps):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps,)),
        keras.layers.Dense(256, activation='relu'),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='rmsprop',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model

```

```python
def train_and_evaluate_models(X_train, y_train, X_val, y_val, X_test, y_test, n_steps, min_val, max_val):
    models = {
        "MLP_1_Baseline": build_mlp_1(n_steps),
        "MLP_2_Regularized": build_mlp_2(n_steps),
        "MLP_3_RMSProp": build_mlp_3(n_steps)
    }

    results = {}

    for name, model in models.items():
        print(f"\nEntrenando {name}...")
        history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                            epochs=50, batch_size=32, verbose=0)

        # Evaluación con keras
        loss, mae, mape, mse = model.evaluate(X_test, y_test, verbose=0)

        # Desnormalizar predicciones y etiquetas para calcular R²
        preds = model.predict(X_test).flatten()
        preds_denorm = preds * (max_val - min_val) + min_val
        y_test_denorm = y_test * (max_val - min_val) + min_val
        r2 = r2_score(y_test_denorm, preds_denorm)

        results[name] = {
            "model": model,
            "history": history,
            "MAE": mae,
            "MAPE": mape,
            "MSE": mse,
            "R2": r2
        }

        print(f"{name} - MAE: {mae:.4f}, MAPE: {mape:.2f}%, MSE: {mse:.4f}, R²: {r2:.4f}")
    
    return results
```

```python
results_mlp = train_and_evaluate_models(
    X_train, y_train,
    X_val, y_val,
    X_test, y_test,
    n_steps, min_val, max_val
)
```

```python
def plot_results(results, X_test, y_test, min_val, max_val):
    plt.figure(figsize=(12, 5))
    
    for name, data in results.items():
        model = data["model"]
        preds = model.predict(X_test).flatten()

        preds_denorm = preds * (max_val - min_val) + min_val
        y_test_denorm = y_test * (max_val - min_val) + min_val

        print(f"{name} - MAE: {data['MAE']:.4f}, MAPE: {data['MAPE']:.2f}%, MSE: {data['MSE']:.4f}, R²: {data['R2']:.4f}")

        if name == list(results.keys())[0]:
            plt.plot(y_test_denorm[:100], label='Real', color='black')
        plt.plot(preds_denorm[:100], label=name, alpha=0.7)

    plt.title("Comparación de Predicciones")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def get_best_model(results):
    best_model = None
    best_mae = float('inf')
    best_name = None

    for name, data in results.items():
        current_mae = data['MAE']
        if current_mae < best_mae:
            best_mae = current_mae
            best_model = data['model']
            best_name = name

    print(f"\nEl mejor modelo es '{best_name}' con MAE: {best_mae:.4f}")
    return best_model, best_name, best_mae
```

```python
def plot_predictions(model, X, y, index, min_val, max_val, n_points=100, title="Predicciones vs Real"):
    """
    Realiza predicciones con el modelo, invierte la normalización y grafica los resultados.
    
    Parámetros:
    - model: Modelo entrenado.
    - X: Datos de entrada (X_train, X_val o X_test).
    - y: Valores reales (normalizados) correspondientes.
    - index: Índice con fechas o etiquetas, que debe tener la misma longitud que y.
    - min_val, max_val: Valores usados en la normalización.
    - n_points: Número de puntos a graficar (por defecto 100).
    - title: Título de la gráfica.
    """
    # Realizar predicciones
    y_pred = model.predict(X).flatten()
    # Invertir la normalización: valor_original = valor_normalizado * (max_val - min_val) + min_val
    y_pred_denorm = y_pred * (max_val - min_val) + min_val
    y_denorm = y * (max_val - min_val) + min_val
    
    # Crear DataFrame para graficar
    df_pred = pd.DataFrame({'Real': y_denorm, 'Predicción': y_pred_denorm}, index=index)
    plt.figure(figsize=(12, 6))
    df_pred.iloc[:n_points].plot(title=title)
    plt.xlabel("Tiempo")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
```

```python
def build_cnn_1(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps, n_features)),  
        keras.layers.Conv1D(64, kernel_size=2, activation='relu'),
        keras.layers.MaxPooling1D(),
        keras.layers.Flatten(),
        keras.layers.Dense(50, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model

def build_cnn_2(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps, n_features)),
        keras.layers.Conv1D(128, kernel_size=3, activation='relu'),
        keras.layers.Conv1D(64, kernel_size=2, activation='relu'),
        keras.layers.MaxPooling1D(),
        keras.layers.Dropout(0.3),
        keras.layers.Flatten(),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model

def build_cnn_3(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Input(shape=(n_steps, n_features)),
        # Bloque 1
        keras.layers.Conv1D(64, kernel_size=3, activation='tanh', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.Conv1D(64, kernel_size=3, activation='tanh', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling1D(pool_size=2),
        keras.layers.Dropout(0.25),
        
        # Bloque 2
        keras.layers.Conv1D(128, kernel_size=3, activation='tanh', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.Conv1D(128, kernel_size=3, activation='tanh', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling1D(pool_size=2),
        keras.layers.Dropout(0.25),
        
        # Capa final
        keras.layers.Flatten(),
        keras.layers.Dense(128, activation='tanh'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(1)
    ])
    
    model.compile(optimizer='adam',
                  loss='mae',
                  metrics=[
                      tf.keras.metrics.MeanAbsoluteError(),
                      tf.keras.metrics.MeanAbsolutePercentageError(),
                      tf.keras.metrics.MeanSquaredError()
                  ])
    return model


```

```python
def train_and_evaluate_cnn_models(X_train, y_train, X_val, y_val, X_test, y_test, n_steps, n_features, min_val, max_val):
    models = {
        "CNN_1_Baseline": build_cnn_1(n_steps, n_features),
        "CNN_2_Deep_Dropout": build_cnn_2(n_steps, n_features),
        "CNN_3_BatchNorm_Tanh": build_cnn_3(n_steps, n_features)
    }

    results = {}

    for name, model in models.items():
        print(f"\nEntrenando {name}...")
        history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                            epochs=50, batch_size=32, verbose=0)

        loss, mae, mape, mse = model.evaluate(X_test, y_test, verbose=0)

        preds = model.predict(X_test).flatten()
        preds_denorm = preds * (max_val - min_val) + min_val
        y_test_denorm = y_test * (max_val - min_val) + min_val
        r2 = r2_score(y_test_denorm, preds_denorm)

        results[name] = {
            "model": model,
            "history": history,
            "MAE": mae,
            "MAPE": mape,
            "MSE": mse,
            "R2": r2
        }

        print(f"{name} - MAE: {mae:.4f}, MAPE: {mape:.2f}%, MSE: {mse:.4f}, R²: {r2:.4f}")
    
    return results
```

```python
def build_lstm_1(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.LSTM(50, activation='tanh', input_shape=(n_steps, n_features)),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam', loss='mae', metrics=[
        tf.keras.metrics.MeanAbsoluteError(),
        tf.keras.metrics.MeanAbsolutePercentageError(),
        tf.keras.metrics.MeanSquaredError()
    ])
    return model

def build_lstm_2(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.LSTM(64, return_sequences=True, input_shape=(n_steps, n_features)),
        keras.layers.Dropout(0.3),
        keras.layers.LSTM(32),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam', loss='mae', metrics=[
        tf.keras.metrics.MeanAbsoluteError(),
        tf.keras.metrics.MeanAbsolutePercentageError(),
        tf.keras.metrics.MeanSquaredError()
    ])
    return model

def build_lstm_3(n_steps, n_features):
    model = keras.Sequential([
        keras.layers.Bidirectional(
            keras.layers.LSTM(50, activation='tanh'), 
            input_shape=(n_steps, n_features)
        ),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam', loss='mae', metrics=[
        tf.keras.metrics.MeanAbsoluteError(),
        tf.keras.metrics.MeanAbsolutePercentageError(),
        tf.keras.metrics.MeanSquaredError()
    ])
    return model
```

```python
def train_and_evaluate_lstm_models(X_train, y_train, X_val, y_val, X_test, y_test, n_steps, n_features, min_val, max_val):
    models = {
        "LSTM_1_Baseline": build_lstm_1(n_steps, n_features),
        "LSTM_2_Stacked_Dropout": build_lstm_2(n_steps, n_features),
        "LSTM_3_Bidirectional": build_lstm_3(n_steps, n_features)
    }

    results = {}

    for name, model in models.items():
        print(f"\nEntrenando {name}...")
        history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                            epochs=50, batch_size=32, verbose=0)

        loss, mae, mape, mse = model.evaluate(X_test, y_test, verbose=0)

        preds = model.predict(X_test).flatten()
        preds_denorm = preds * (max_val - min_val) + min_val
        y_test_denorm = y_test * (max_val - min_val) + min_val
        r2 = r2_score(y_test_denorm, preds_denorm)

        results[name] = {
            "model": model,
            "history": history,
            "MAE": mae,
            "MAPE": mape,
            "MSE": mse,
            "R2": r2
        }

        print(f"{name} - MAE: {mae:.4f}, MAPE: {mape:.2f}%, MSE: {mse:.4f}, R²: {r2:.4f}")
    
    return results
```
