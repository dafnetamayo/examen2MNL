```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_series = pd.read_csv('Hourly-train.csv')
df_meta = pd.read_csv('m4_info.csv')

print(df_series.head())
print(df_meta.head())
```

```python
#extraer y estandarizar el valor de la columna 'SP' de la primera fila
sp_value = df_meta.loc[0, 'SP']
sp_value_lower = sp_value.lower()
```

```python
#Preparación del DataFrame de series
# Se establece la primera columna como índice para organizar mejor el DataFrame
df_series.set_index(df_series.columns[0], inplace=True)

# Transponemos el DataFrame para que cada columna represente una serie de tiempo
df_t = df_series.T
```

```python
series = ['H110', 'H7', 'H395', 'H183', 'H49'] 
filtered = df_meta.loc[df_meta['M4id'].isin(series)]
filtered
```

# serie H110

```python
serie_id = "H110"

start_date = pd.to_datetime("01-07-15 12:00", format="%d-%m-%y %H:%M")
# Genera el índice de fechas con frecuencia horaria, con tantos periodos como datos tenga la serie
index = pd.date_range(start=start_date, periods=len(df_t[serie_id]), freq="H")

serie = df_t[serie_id].copy()
serie.index = index
```

```python
df_serie = pd.DataFrame({
    "fecha": index,
    "valor": serie.values
})
#eliminar filas con datos faltantes
df_serie = df_serie.dropna(subset=['valor'])
```

```python
# Visualización de la serie original
plt.figure(figsize=(8, 4))
plt.plot(df_serie["fecha"], df_serie["valor"], label=serie_id)
plt.xlabel("Fecha y Hora")
plt.ylabel("Valor")
plt.title(f"Evolución de la serie {serie_id}")
plt.legend()
plt.show()
```

```python
#transformación: Aplicar transformación logarítmica
# Se suma 1 para evitar problemas con log(0)
df_serie["valor_log"] = np.log(df_serie["valor"] + 1)

plt.figure(figsize=(8, 4))
plt.plot(df_serie["fecha"], df_serie["valor_log"], label=f"{serie_id} (log-transformed)")
plt.xlabel("Fecha y Hora")
plt.ylabel("Valor Logarítmico")
plt.title(f"Evolución de la serie {serie_id} (Transformación Logarítmica)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

def plot_log_decomposition_from_id(df_series, df_meta, series_id, period=24):
    """
    Toma una serie como 'H110', encuentra su columna, aplica log +1 y muestra su descomposición.
    """
    # Buscar metadata de la serie
    row = df_meta[df_meta["M4id"] == series_id]
    if row.empty:
        print(f"Serie no encontrada")
        return
    
    # Obtener columna real (Vxxx)
    v_col = "V" + str(int(series_id[1:]) + 1)  # H110 -> V111
    if v_col not in df_series.columns:
        print(f"Columna no existe ")
        return

    # Obtener fecha de inicio
    start_date_str = row.iloc[0]["StartingDate"]
    start_date = pd.to_datetime(start_date_str, format="%d-%m-%y %H:%M")

    # Extraer serie, asignar fechas y eliminar NaN
    serie = df_series[v_col].dropna()
    index = pd.date_range(start=start_date, periods=len(serie), freq="H")
    serie.index = index

    # Aplicar log(1+x)
    serie_log = np.log(serie + 1)

    # Descomposición
    decomposition = seasonal_decompose(serie_log, model='additive', period=period)

    # Graficar
    fig, axes = plt.subplots(4, 1, figsize=(15, 12))
    titles = ["Serie Logarítmica", "Tendencia", "Estacionalidad", "Residuo"]
    components = [decomposition.observed, decomposition.trend, decomposition.seasonal, decomposition.resid]

    for ax, title, comp in zip(axes, titles, components):
        ax.plot(comp)
        ax.set_title(f"{title} - {series_id}")
        ax.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
series_list = ['H110', 'H7', 'H395', 'H183', 'H49']

for sid in series_list:
    plot_log_decomposition_from_id(df_series, df_meta, sid)
```
